```python
# Set seaborn style with grid lines
sns.set(style='darkgrid', rc={'grid.linestyle': '--'})

# Add column to mark max value and get max month per year
def mark_max_and_get_legend(group):
    max_row = group.loc[group['passengers'].idxmax()]
    group['highlight'] = group['passengers'] == max_row['passengers']
    group['max_info'] = f" {max_row['month']}: {int(max_row['passengers'])} "
    return group

flights = flights.groupby('year').apply(mark_max_and_get_legend).reset_index(drop=True)

# Set month order to keep x-axis consistent
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create FacetGrid by year
g = sns.FacetGrid(flights, col='year', col_wrap=4, sharey=True, height=3.5)

# Map barplot to each subplot
g.map_dataframe(
    sns.barplot,
    x='month',
    y='passengers',
    hue='highlight',
    palette={True: '#c71a3a', False: '#2a6fde'},
    hue_order=[False, True],
    order=month_order
)

# Rotate x-axis labels, enable grid, and add custom legend per subplot
for ax, (year, group) in zip(g.axes.flatten(), flights.groupby('year')):
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle='--',color='gray', alpha=0.5)
    
    # Get the max month info string for legend
    legend_label = group['max_info'].iloc[0]
    
    # Place legend as a text box
    ax.text(0.01, 0.98, legend_label,
            transform=ax.transAxes,
            ha='left', va='top',
            fontsize=9,
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))



# Set titles and labels
g.set_titles("Year {col_name}")
g.set_axis_labels("Month", "Passengers")

plt.tight_layout()
plt.show()
```
