# Passengers v/s year

```python
plt.figure(figsize=(10,6))
plt.title("Passenger v/s Year")
plt.grid(True, linestyle='--', color='grey', alpha = 0.5)
sns.lineplot(data= flights, x='year', y='passengers')
```

```python
sns.scatterplot(data=flights, x='year', y='passengers')
```

```python
sns.displot(data=flights['passengers'], kind='kde')
```

```python
sns.violinplot(data=flights, x='month', y='passengers')
```

```python
sns.barplot(data=flights, x='month', y='passengers')
```

```python
flights.groupby('month')['passengers'].mean()
```

```python
sns.lmplot(data=flights, x='year', y='passengers')
```

```python
flights['year']=='1960'
```

```python
# prompt: i wanted to analyse months v/s passengers in year==1960

# # Months v/s passengers in year==1960
plt.figure(figsize=(10,6))
plt.title("Months v/s passengers in year==1960")
plt.grid(True, linestyle='--', color='grey', alpha = 0.5)
sns.lineplot(data= flights[flights['year']==1960], x='month', y='passengers')
sns.scatterplot(data=flights[flights['year']==1960], x='month', y='passengers')
sns.violinplot(data=flights[flights['year']==1960], x='month', y='passengers')
sns.barplot(data=flights[flights['year']==1960], x='month', y='passengers')
flights[flights['year']==1960].groupby('month')['passengers'].mean()
```
